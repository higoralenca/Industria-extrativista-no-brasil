```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

# Produção

- Carvão mineral

```python
# PIB em milhões - 2010 / 2020

tabelaPIB_CarvãoMineral = pd.read_csv("Extrativismo - Carvão Mineral - Produção - PIB.csv")
tabelaPIB_CarvãoMineral.rename(columns={"05 Extração de carvão mineral- Produção- PIB - Extração de carvão mineral": "PIB"}, inplace=True)
display(tabelaPIB_CarvãoMineral)
```

```python
sns.lineplot(tabelaPIB_CarvãoMineral, x= "Ano", y="PIB", color="black")
```

```python
# Selecionando X e Y

y = tabelaPIB_CarvãoMineral["PIB"]
X = tabelaPIB_CarvãoMineral.drop("PIB", axis=1)

# Importando o train_test_split

from sklearn.model_selection import train_test_split

# Separando essa base em treino e teste

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Importando a regressão linear

from sklearn.linear_model import LinearRegression

# Criando o regressor e fazendo o fit com os dados de treino

reg_rl = LinearRegression().fit(X_train, y_train)
```

```python
# Fazendo a previsão pros dados de teste

Teste = tabelaPIB_CarvãoMineral.drop("PIB", axis=1)

reg_rl.predict(Teste)
```
